### Bài 2
Viết một chương trình tìm tất cả các số trong đoạn 1000 và 3000 (tính cả 2 số này) sao cho tất cả các chữ số trong số đó là số chẵn. In các số tìm được thành chuỗi cách nhau bởi dấu phẩy, trên một dòng.

```python
lst2=[str(i) for i in range(1000,3001) if i%2==0]
st= ",".join(lst2)
print(st)
```

### Bài 3
Viết một chương trình chấp nhận đầu vào là một câu, đếm số chữ cái và chữ số trong câu đó. 

Chẳng hạn, chuỗi được nhập vào là: 

> hello world! 123

Thì đầu ra sẽ là:

> Số chữ cái là: 10
>
> Số chữ số là: 3


```python
a= input("Nhập chuỗi:")
count1=count2=0
for i in a:
    if i.isdigit() :
        count1+=1
    elif i.isalpha():
        count2+=1
print("Số chữ số là:",count1,"\nSố chữ cái là:",count2)
```

### Bài 4
Viết một chương trình thực hiện các thao tác sau:
* Nhập vào từ bàn phím Họ Tên, và Điểm thi môn lập trình Python của sinh viên. Quá trình nhập sẽ dừng khi không có Họ Tên mới được nhập. 
* Dữ liệu được ghi vào một từ điển, với Key=Họ Tên, và Value=Điểm. 
* Sắp xếp danh sách theo thứ tự tăng dần của Điểm.

```python
dct={}
while True:
    a= input("Nhập họ tên:")
    if a=="":
        break
    b= float(input("Nhập điểm môn Python:"))
    dct[a]=b
dct= {k: v for k, v in sorted(dct.items(), key=lambda item: item[1])}
print(dct)
```

### Bài 5
Tính xấp xỉ tích phân xác định sau đây bằng tổng Riemann:
$$I=\int^1_0 \dfrac{x}{1+x^2}dx$$
* Định nghĩa hàm $f(x)=\dfrac{x}{1+x^2}$
* Chia đoạn lấy tích phân [0,1] thành $n=10$ đoạn con bằng nhau với độ rộng $\Delta x=\dfrac{1}{n}$
* Tính tổng Riemann: 
$$sum=\sum^n_{i=1}f(x_i)\times\Delta x$$
với $x_i$ là điểm giữa của khoảng chia thứ i.
* Tăng số khoảng chia $n$ và nhận xét về kết quả.

```python
#Khoảng chia 10
n=int(input("Nhập số khoảng:"))
def f(x):
    return x/(1+x**2)
deltax= 1/n
sum_riemann=0
xi=deltax/2
for step in range(n):
    sum_riemann += f(xi)*deltax
    xi += deltax
print(sum_riemann)
```

***Nhận xét: Khoảng chia n càng lớn thì tổng Riemann càng chính xác***
