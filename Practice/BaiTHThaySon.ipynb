```python
#Bài TH3
from random import randint
list=["Búa","Kéo","Giấy"]
while True:
    a=input("Nhập lựa chọn:")
    b=list[randint(0,2)]
    print("Máy chọn:",b)
    if a==b:
        print("Hòa rồi")
    else:
        if a == "Giấy":
            if b == "Búa":
                print("Thắng rồi!")
            else:
                print("Thua")
        elif a == "Búa":
            if b == "Kéo":
                print("Thắng rồi!")
            else:
                print("Thua")
        elif a == "Kéo":
            if b == "Giấy":
                print("Thắng rồi!")
            else:
                print("Thua")
        else:
            print("Nhập sai!")
    print("*"*30)
    c=input("Chơi tiếp k? (c/k):")
    if c=="k":
        break
```

```python
#Bài TH4
seconds=int(input("Nhập số giây hoàn thành công việc (s):"))
a="AM" if seconds<43200 else "PM"

hours= seconds//3600 if seconds<43200 else (seconds-43200)//3600

minutes=(seconds%3600)//60
seconds=(seconds%3600)-minutes*60

minutes= "0"+str(minutes) if minutes<10 else minutes
seconds="0"+str(seconds) if seconds<10 else seconds
print(f"Thời điểm hoàn thành công việc {hours}:{minutes}:{seconds} {a}")
```

```python
#Bài TH5 
from math import sqrt
count=0
for i in range(1,31):
    for j in range(1,31):
        s=sqrt(i**2+j**2)
        if i+j>s and i+s>j and j+s>i and s%1 == 0:
            count+=1
            print(f"Tam giác {count} có 2 cạnh góc vuông và cạnh huyền lần lượt là:",i,j,int(s),sep="\t")
print("Tổng số tam giác thỏa mãn:",count)
```

```python
# Bài 5.1
# tạo 1 set , p=2 xóa bội của p k xóa p, p=p+1
lim=float(input("Nhập giới hạn:"))
lim= int(lim)
set1=set(i for i in range(2,lim+1))
print(set1)
p=2
while p in set1:
    set2=set()
    for i in set1:
        if i%p==0 and i>p:
            set2.add(i)
    set1= set1- set2
    print(p)
    if p==max(set1):
        break
    for i in range(p,lim+1):
        if i in set1 and i>p:
            p=i
            break
print("Tập các số nguyên tố:",sorted(set1),sep="\n")
```

```python
#Nhập dữ liệu mua:
dulieumua = list()
while True:
    line = input("Nhập dữ liệu mua: ")
    if line == '':
        break
    else:
        line = line.split(',')
        line[0] = int(line[0])
        line[1] = int(line[1])
        line[2] = int(line[2])
        line[3] = int(line[3])
        dulieumua.append(tuple(line))
print("Dữ liệu mua", dulieumua)
tongtien = 0
for x in dulieumua:
    if x[4] == 0:
        tongtien = (x[1]*300000*0.3) + (x[2]*300000*0.2) + ((x[0]-x[1]-x[2])*300000)
        dulieumua.append(tongtien)
    elif x[4] == 1:
        tongtien = (x[1]*300000*0.3) + (x[2]*300000*0.2) + ((x[0]-x[1]-x[2])*300000)*0.05
        dulieumua.append(tongtien)
# in dữ liệu:
for x in dulieumua:
    print("Tổng số vé mua", x[0])
    print("Số vé giảm 30%",x[1])
    print("Số vé giảm 20%",x[2])
    print("Chiết khấu 5%",x[3])
    print("Tổng số tiền phải trả",x[4])
```
