# Bài TH-4-3: Emirp
Bạn hãy viết một chương trình tìm tất cả các số <b>emirp</b> nhỏ hơn 10 vạn (100000).
<blockquote>
    <ul>
        <li>Nếu bạn đang cố tra từ trên trong tiếng anh thì chắc sẽ không tìm thấy đâu, bởi nó là từ viết ngược của từ <b>“Prime - Số nguyên tố”.</b></li>
    <li>Một <b>emirp</b> là một số nguyên tố mà khi đảo ngược vị trí các chữ số của nó, ta cũng được một số nguyên tố.<br>
Định nghĩa này không bao gồm các số nguyên tố xuôi ngược (như 151 hoặc 787), cũng không phải số nguyên tố  có một chữ số như số 7.</li>
    <li><b>Ví dụ:</b> Số 13 là một số <b>emirp</b> vì đảo ngược của nó là số 31 cũng là số nguyên tố. Đương nhiên, số 31 cũng là một số <b>emirp</b>.</li>
    </ul>
</blockquote>

```python
def is_prime(n):
    count = 0
    for i in range(1, int(n/2)):
        if n % i == 0:
            count += 1
    if count == 1:
        return True
    return False
for i in range(11,100000,2):
    _i= int(str(i)[::-1])
    if is_prime(i) ==is_prime(_i)==True:
        print(i)
```

# Bài TH-4-4: Cặp số thân thiết
Bạn hãy viết một chương trình tìm ra các <b>cặp số thân thiết</b> với giá trị các số nhỏ hơn 1 vạn (10000).
<blockquote><b>Biết rằng:</b> Hai số tạo thành một cặp số thân thiết khi số này bằng tổng tất cả các ước thực sự của số kia và ngược lại.<br>
    <b>Lưu ý:</b> Sau khi thông báo cặp số thân thiết, cần đưa ra minh chứng.<br>
    <b>Ví dụ:</b> Cặp (220, 284) là một cặp số thân thiết, vì:
    <ul>
        <li>Tổng các ước thực sự của 220 là: 1+2+4+5+10+11+20+22+44+55+110 = 284.</li>
        <li>Tổng các ước thực sự của 284 là: 1+2+4+71+142 = 220.
    </ul>
</blockquote>

```python
def tong_uoc(x):
    result=0
    for i in range(1,x//2+1):
        if x%i==0:
            result+=i
    return result
tuple_lst=[]
for i in range(1,10001):
    x_factor= tong_uoc(i)
    if tong_uoc(x_factor)==i and i != x_factor and (x_factor,i) not in tuple_lst:
        tuple_lst.append((i,x_factor))
        print(i,x_factor)
```

```python
print(tong_uoc(2620))
```

```python
diction={k:v for k in KEY for v in VALUE}
```
