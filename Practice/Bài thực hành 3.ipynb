### Bài 1
Một từ, hoặc số được gọi là **Panlindrome** nếu đọc xuôi hoặc ngược đều giống nhau, chẳng hạn như từ *madam*, *level*, hay ngày *11.11.11*.
* Hãy viết một chương trình nhập một chuỗi từ người dùng và sử dụng vòng lặp để kiểm tra xem chuỗi có phải là chuỗi *Panlindrome* hay không?
* Viết một hàm đệ quy thực hiện việc kiểm tra một chuỗi có là *Panlindrome* hay không theo hướng dẫn dưới đây:
> **Hướng dẫn**: 
> * Chuỗi rỗng là *Panlindrome*, 
> * Chuỗi có 1 phần tử là *Panlindrome*
> * Chuỗi là *Panlindrome* nếu hai kí tự đầu, cuối giống nhau, và chuỗi con sau khi bỏ đi 2 kí tự đầu, cuối cũng là *Panlindrome*

```python
def ispanlindrome(st):
    if len(st)==0 or len(st)==1:
        return True
    elif st[0]==st[-1]:
        return ispanlindrome(st[1:len(st)-1])
    else:
        return False
x= input("Nhập chuỗi để kiểm tra:")
check = ispanlindrome(x)
tx = "có" if check ==True else "không"
print(f"{x} {tx} là chuỗi Panlindrome")
```

### Bài 2
Phân tích ra thừa số nguyên tố của một số nguyên, $n$, có thể được xác định sử dụng thuật toán sau đây:

>> 

Viết một chương trình nhập một số nguyên bởi người dùng. Nếu số đó nhỏ hơn 2 thì báo lỗi. Nếu không, hãy hiển thị các số nguyên tố trong phân tích của số được nhập.

```python
while True:
    n=int(input("Nhập n:"))
    if 2<=n:
        break
    else:
        print("Lỗi: n không hợp lệ vui lòng nhập lại.")
factor = 2
while factor <= n:
    if n % factor == 0:
        print(f"{factor} là thừa số của {n}")
        n //=factor
    else:
        factor+=1
```

### Bài 3
Viết một chương trình sử dụng bộ sinh số ngẫu nhiên để mô phỏng quá trình gieo đồng xu một số lần. 
* Đồng xu phải cân đối và đồng chất, nghĩa là xác suất gieo được mặt sấp hay mặt ngửa là như nhau. 
* Chương trình sẽ tự động gieo đồng xu cho đến khi 3 lần liên tiếp xảy ra cùng một kết quả, chẳng hạn gieo được 3 mặt ngửa hay 3 mặt sấp, thì dừng. 
* Hiển thị các kết quả gieo trên cùng 1 dòng, mặt ngửa kí hiệu là H và mặt sấp kí hiệu là T. 
* Hiển thị số lần gieo cho đến khi dừng. 
* Chương trình sẽ mô phỏng 10 lần gieo đồng xu và tính số lần gieo trung bình.
> **Minh họa**:
> 


```python
def coin_flip():
    from random import randint
    k = ["H","T"]
    count=count_loop=0
    lst=[]
    while True:
        count+=1
        lst.append(k[randint(0,1)])
        if len(lst)==1:
            continue
        elif lst[-1]==lst[-2]:
            count_loop+=1
        else:
            count_loop=0
        if count_loop==2:
            break
    return " ".join(lst) +f" ( {count} flips)"
```

```python
avg=[]
for i in range(10):
    x= coin_flip()
    avg.append(int(x[-9:-7]))
    print(x)
print("On average, ",sum(avg)/len(avg)," flips were needed")
```

### Bài 4
Khoảng cách **Edit Distance (ED)** hay còn gọi là khoảng cách **Levenshtein** là số nhỏ nhất các thao tác: 
* chèn một kí tự (letter insertion), 
* xóa kí tự (letter deletion), và 
* đổi kí tự (letter substitution) 

để biến chuỗi này thành chuỗi kia. Ví dụ 2 chuỗi *kitten* và *sitting* có khoảng cách ED bằng 3 vì:
> **k**itten --> sitt**e**n --> sittin --> sittin**g**

Viết một hàm đệ quy tính khoảng cách ED giữa hai chuỗi, sử dụng thuật toán sau đây:
> 

```python
def edit_distance(st1,st2):
    if len(st1) == 0:
        return len(st2)
    elif len(st2) == 0:
        return len(st1)
    else:
        cost=0
        if st1[-1] != st2[-1]:
            cost=1
        d1 = edit_distance(st1[:len(st1)-1],st2)+1
        d2 = edit_distance(st1,st2[:len(st2)-1]) + 1
        d3 = edit_distance(st1[:len(st1)-1],st2[:len(st2)-1]) + cost
        return min([d1,d2,d3])
```

```python
edit_distance("kitten","sitting")
```
