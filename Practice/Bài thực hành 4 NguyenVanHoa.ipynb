### Bài 1
* Viết một chương trình nhập vào 2 số nguyên và sử dụng thuật toán dưới đây để tìm ước số chung lớn nhất của hai số đó:

> 

* Viết một hàm nhận hai số nguyên biểu diễn tử số và mẫu số của một phân số như là tham số của nó. Hàm trả về một cặp số nguyên tương ứng là phân số đã được rút gọn.

> **Ví dụ**: *frac_reduced*(6,21) = (2,7)

```python
def UCLN(m,n):
    d = m if m < n else n
    while m % d != 0 or n % d != 0:
        d -= 1
    return d

print(UCLN(6,21))

def frac_reduced(a,b):
    t = UCLN(a,b)
    return (int(a/t),int(b/t))

frac_reduced(6,21)
```

### Bài 3
Viết một chương trình thực hiện:
* Nhập vào các cặp số thực $(x_i, y_i)$ từ bàn phím, bỏ trống khi muốn dừng nhập dữ liệu.
* In ra danh sách dữ liệu.
> **Ví dụ**:
>> Nhập dữ liệu (thoát khi bỏ trống): 1,1
>>
>> Nhập dữ liệu (thoát khi bỏ trống): 2,2.1
>>
>> Nhập dữ liệu (thoát khi bỏ trống): 3,2.9
>>
>> Nhập dữ liệu (thoát khi bỏ trống): 4,4
>>
>> Nhập dữ liệu (thoát khi bỏ trống):
>>
> Dữ liệu: [(1,1), (2,2.1), (3,2.9), (4,4)]

* Tính và đưa ra đường hồi quy tuyến tính phù hợp nhất với dữ liệu.

```python
data=[]
while True:
    try:
        a=input("Nhập (x,y):")
        if a=="":
            break
        a = tuple(float(x) for x in a.split(","))
        data.append(a)
    except:
        print("Nhập lại")
n=len(data)
xi=[i[0] for i in data]
yi=[i[1] for i in data]
xiyi=[i[0]*i[1] for i in data]
xi2=[i[0]**2 for i in data]
Sxy=sum(xiyi)-sum(xi)*sum(yi)/n
Sxx=sum(xi2)-(sum(xi)**2)/n
B1=Sxy/Sxx
B0=sum(yi)/n-sum(xi)/n*B1
print("Dữ liệu:",data)
print("Đường hồi quy tuyến tính: y={}x+{}".format(round(B1,6),round(B0,6)))
```

### Bài 4
File **'salary_data.txt'** chứa dữ liệu về số năm kinh nghiệm và mức lương của các quan sát khác nhau. Dữ liệu được lưu trữ theo từng dòng. Hãy viết một chương trình thực hiện các công việc sau:
> * Đọc file, và tải dữ liệu vào 2 mảng số (danh sách), 1 mảng tên 'experience' và 1 mảng tên 'salary'.
> * Mô tả thống kê 2 mảng dữ liệu: tính và hiển thị giá trị trung bình, độ lệch tiêu chuẩn, giá trị lớn nhất, nhỏ nhất.
> * Xây dựng hàm hồi quy tuyến tính mô tả lương (y) theo số năm kinh nghiệm (x):
>
> $$y=b0+b1.x$$
>
> * Vẽ đồ thị dữ liệu và đường hồi quy

```python
file = open('salary_data.txt')
data= file.read().split("\n")
data=[i.split(",") for i in data]
data= data[1:len(data)-1]
experience = [ float(i[0]) for i in data]
salary = [ float(i[1]) for i in data]
def data_describe(data):
    from math import sqrt
    n= len(data)
    x_tb = sum(data)/n 
    x2 = [i**2 for i in data]  
    x2_tb = sum(x2)/n  
    std = sqrt(n/(n-1)*(x2_tb-x_tb**2))
    return [x_tb,std,max(data),min(data)]
exp_des= data_describe(experience)
sal_des= data_describe(salary)
print(f"Mô tả thống kê Experience:\n\tGiá trị trung bình: {exp_des[0]}\t\
Độ lệch tiêu chuẩn: {exp_des[1]}\tGTLN: {exp_des[2]}\tGTNN: {exp_des[3]}")
print(f"Mô tả thống kê Salary:\n\tGiá trị trung bình: {sal_des[0]}\t\
Độ lệch tiêu chuẩn: {sal_des[1]}\tGTLN: {sal_des[2]}\tGTNN: {sal_des[3]}")
```

```python
#Đường hồi quy
n= len(experience)
sum_xi=sum_yi=sum_xiyi=sum_xi2=0
for i in range(n):
    sum_xi += experience[i]
    sum_yi += salary[i]
    sum_xiyi += experience[i]*salary[i]
    sum_xi2 += experience[i]**2
Sxy=sum_xiyi-sum_xi*sum_yi/n
Sxx=sum_xi2-(sum_xi**2)/n
B1=Sxy/Sxx
B0=sum_yi/n-sum_xi/n*B1
print("Đường hồi quy tuyến tính: y={}x+{}".format(round(B1,6),round(B0,6)))
```
